```
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
%matplotlib inline
```

```
image = mpimg.imread('curved_lane.jpg')
plt.imshow(image)
```

```
import cv2
```

```
# convert image from RGB colorspace to GRAY colorspace
gray_image = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
plt.imshow(gray_image, cmap='gray')
```

## Now applying filters/kernels to the given image

```
import numpy as np
```

```
# custom kernel (to blur out the image) - takes an average of pixels
blur_filter = (1/9)*np.array([[1,1,1],
                          [1,1,1],
                          [1,1,1]])

blur_image = cv2.filter2D(gray_image,-1,blur_filter)

plt.imshow(blur_image, cmap='gray')
```

```
# 3 by 3 filter (an array) for edge detection
sobel_y = np.array([[1,2,1],
                    [0,0,0],
                    [-1,-2,-1]])

# Filter the image using Filter2D function
filtered_image = cv2.filter2D(blur_image, -1, sobel_y) # ddepth = -1, means the output image after applying the filter has same depth as the input image

plt.imshow(filtered_image, cmap='gray')
```

```
# 3 by 3 filter (an array) for edge detection
sobel_x = np.array([[1,0,-1],
                    [2,0,-2],
                    [1,0,-1]])

# Filter the image using Filter2D function
filtered_image = cv2.filter2D(blur_image, -1, sobel_x) # ddepth = -1, means the output image after applying the filter has same depth as the input image

plt.imshow(filtered_image, cmap='gray')
```
