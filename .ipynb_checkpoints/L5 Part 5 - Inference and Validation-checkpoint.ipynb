# Inference and Validation

__Inference__ : making predictions using the trained network(model)

__Validation__ : Checking the performance of model on data not seen before

__Aim__ : is to minimise the Validation loss (not the training loss)

Only minimising the training loss can result in overfitting.

So while training, we keep a check on the performance of the network on data not in the training set (called Validation set), to __check overfitting__ (__Validation__)

Ways to __reduce overfitting__ :
1. __Early Stopping__ : stopping on a threshold of epochs - so that overfitting does not occur
2. __Dropout__ : randomly dropping out input features - so that the model does not overfit the training data

```python
# Loading the Fashion-MNIST data using torchvision library

import torch
from torchvision import datasets, transforms

# Define a transform to normalize the data
transform = transforms.Compose([transforms.ToTensor(),
                                transforms.Normalize((0.5,), (0.5,))])
# Download and load the training data
trainset = datasets.FashionMNIST('~/.pytorch/F_MNIST_data/', download=True, train=True, transform=transform)
trainloader = torch.utils.data.DataLoader(trainset, batch_size=64, shuffle=True)

# Download and load the test data
testset = datasets.FashionMNIST('~/.pytorch/F_MNIST_data/', download=True, train=False, transform=transform)
testloader = torch.utils.data.DataLoader(testset, batch_size=64, shuffle=True)
```

```python
# Checking the shape of data

images,labels = next(iter(trainloader))

print(images.shape)
print(labels.shape)
```

```python
# Defining a model to train on this data

from torch import nn, optim
import torch.nn.functional as F

class Classifier(nn.Module):
    
    def __init__(self):
        
        super().__init__()
        
        # 3 hidden layers, 1 output layer
        self.fc1 = nn.Linear(784,256)
        self.fc2 = nn.Linear(256,128)
        self.fc3 = nn.Linear(128,64)
        self.fc4 = nn.Linear(64,10)
        
    def forward(self, x):
        
        # must flatten the input tensor
        x = x.view(x.shape[0], -1)
        
        x = F.relu(self.fc1(x))
        x = F.relu(self.fc2(x))
        x = F.relu(self.fc3(x))
        x = F.log_softmax(self.fc4(x), dim=1)
        
        return x
```

```python
# Creating the model and training it

model = Classifier()

criterion = nn.NLLLoss()
optimizer = optim.Adam(model.parameters(), lr = 0.003) # Similar to SGD optimizer, but is better

epochs = 30

train_losses = []
test_losses = []

for e in range(epochs):
    
    training_loss_per_epoch = 0
    test_loss_per_epoch = 0
    accuracy = 0
    
    # Training on Training set
    for images, labels in trainloader:  # (images, labels): batch of 64
        
        logps = model.forward(images)
        loss = criterion(logps,labels)
        
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
        
        training_loss_per_epoch += loss.item()
        
    else:
        
        # Validation using Test set
        # turn off gradients while validation (since we are not training), increases speed and decreases memory 
        with torch.no_grad():
            
            for images, labels in testloader:

                logps = model(images)
                test_loss_per_epoch += criterion(logps, labels)
                
                # cummulative accuracy over all batches of the test set
                ps = torch.exp(logps)
                top_p, top_class = ps.topk(k=1, dim=1)
                equals = top_class == labels.view(top_class.shape)
                accuracy += torch.mean(equals.type(torch.FloatTensor))
              
        train_losses.append(training_loss_per_epoch/len(trainloader))
        test_losses.append(test_loss_per_epoch/len(testloader))
        
        print("Epoch: ", e+1, "  Training loss: ", training_loss_per_epoch/len(trainloader), 
              "  Test loss: ", test_loss_per_epoch/len(testloader), 
              "  Test Accuracy: ", accuracy/len(testloader) )
        
        
```

```python
# Plotting how the training and validation (test) losses vary while training the model/network

%matplotlib inline
%config InlineBackend.figure_format = 'retina'

import matplotlib.pyplot as plt

plt.plot(train_losses, label='Training loss')
plt.plot(test_losses, label='Validation loss')
plt.legend(frameon=False)
```

## Using __Dropout__ : to reduce Overfitting

```python
# Defining a new model to train : using dropout

from torch import nn, optim
import torch.nn.functional as F

class Classifier(nn.Module):
    
    def __init__(self):
        
        super().__init__()
        
        # 3 hidden layers, 1 output layer
        self.fc1 = nn.Linear(784,256)
        self.fc2 = nn.Linear(256,128)
        self.fc3 = nn.Linear(128,64)
        self.fc4 = nn.Linear(64,10)
        
        # to reduce overfitting
        self.dropout = nn.Dropout(p=0.2)
        
    def forward(self, x):
        
        # must flatten the input tensor
        x = x.view(x.shape[0], -1)
        
        x = self.dropout(F.relu(self.fc1(x)))
        x = self.dropout(F.relu(self.fc2(x)))
        x = self.dropout(F.relu(self.fc3(x)))
        x = F.log_softmax(self.fc4(x), dim=1)
        
        return x
```

```python
# Training this new model on the train data, and simultaneously performing validation after every epoch

model = Classifier()

criterion = nn.NLLLoss()
optimizer = optim.Adam(model.parameters(), lr = 0.003)

epochs = 30

train_losses = []
test_losses = []

for e in range(epochs):
    
    training_loss_per_epoch = 0
    test_loss_per_epoch = 0
    accuracy = 0
    
    # Training on Training set
    for images, labels in trainloader:  # (images, labels): batch of 64
        
        logps = model.forward(images)
        loss = criterion(logps,labels)
        
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
        
        training_loss_per_epoch += loss.item()
        
    else:
        
        # Validation using Test set
        
        # turn off gradients while validation (since we are not training), increases speed and decreases memory 
        with torch.no_grad():
            
            model.eval()  # turns off dropout
            
            for images, labels in testloader:

                logps = model(images)
                test_loss_per_epoch += criterion(logps, labels)
                
                # cummulative accuracy over all batches of the test set
                ps = torch.exp(logps)
                top_p, top_class = ps.topk(k=1, dim=1)
                equals = top_class == labels.view(top_class.shape)
                accuracy += torch.mean(equals.type(torch.FloatTensor))
              
        train_losses.append(training_loss_per_epoch/len(trainloader))
        test_losses.append(test_loss_per_epoch/len(testloader))
        
        print("Epoch: ", e+1, "  Training loss: ", training_loss_per_epoch/len(trainloader), 
              "  Test loss: ", test_loss_per_epoch/len(testloader), 
              "  Test Accuracy: ", accuracy/len(testloader) )
        
```

```python
# Plot : Training loss vs Validation/Test Loss (when using dropout technique)

%matplotlib inline
%config InlineBackend.figure_format = 'retina'

import matplotlib.pyplot as plt

plt.plot(train_losses, label='Training loss')
plt.plot(test_losses, label='Validation loss')
plt.legend(frameon=False)
```

Better. Overfitting reduced - both training and validation loss decreasing!

Not as in the model without dropout - where the training loss kept decreasing, but the validation loss was increasing.

## Inference

Now that the model is trained, we can use it for inference. We've done this before, but now we need to remember to set the model in inference mode with `model.eval()` - __to turn off dropout__. You'll also want to __turn off autograd__ with the `torch.no_grad()` context.

```python
# Setting the trained model to evaluation mode - to turn off dropout - since we are not training
model.eval() # turn off dropouts

# grabbing a batch of test images
images, labels = next(iter(testloader))

# Picking one image
img = images[0]

# Using the trained model/network (flattening of input - done within the model)

with torch.no_grad(): # turnn off the autograd - since no training is done here
    log_ps = model(img)

ps = torch.exp(log_ps)

# The test image and the predicted class probabilities
import helper
helper.view_classify(img.view(28,28),ps,version='Fashion')
print(labels[0])

# Set the model back to training mode
model.train()
```
